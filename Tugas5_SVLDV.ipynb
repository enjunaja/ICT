````python
# Membuat Menu Simulasi
def menu_simulasi():
    clear_output()
    print('Ingat kembali Bentuk Umum SPLDV')
    print('ax + by = c')
    print('px + qy = r')
    time.sleep(5)
    print('')
    print('dengan:')
    print('x dan y adalah variabel')
    print('a, b, p, dan q adalah koefisien')
    print('c dan r adalah konstanta')
    print('Nilai x dan y yang memenuhi kedua persamaan tersebut adalah solusi SPLDV')
    time.sleep(10)
    print('')

    print("Pada menu ini, teman-teman akan diminta untuk")
    time.sleep(2)
    print('menginputkan nilai koefisien-koefisien, yaitu a, b, p, dan q')
    time.sleep(2)
    print('dan konstanta-konstanta-konstanta, yaitu c dan r')
    time.sleep(2)
    print('dari suatu SPLDV')
    time.sleep(2)
    print("kemudian program akan memberikan solusi dari SPLDV yang teman-teman buat.")
    print('Sudah siap ya!')
    time.sleep(10)
    print('')

    #Masukan Persamaan 1
    clear_output(wait=True)
    print('Oke, sekarang kita mulai!')
    print('Dari SPLDV yang teman-teman buat, pastikan formatnya')
    print('sudah sesuai dengan Bentuk Umum SPLDV')
    time.sleep(5)
    print('')
    print('Persamaan 1: ax + by = c')
    print('Persamaan 2: px + qy = r')
    time.sleep(10)
    print('')

    print('Dari persamaan 1, inputkanlah nilai-nilai koefisien (a dan b) dan konstantanya c)!')
    time.sleep(3)
    print('Ingat bahwa: Nilai a',"\u2260",'0 atau b',"\u2260",'0')
    time.sleep(3)
    print('Jika nilai a, b, atau c adalah bilangan pecahan')
    time.sleep(3)
    print('Teman-teman bisa inputkan dalam bentuk desimalnya')
    time.sleep(3)
    print('atau diinputkan dalam bentuk pecahan "m/n"')
    time.sleep(10)
    print('')

    Kondisiab = False
    while Kondisiab == False:

     Kondisia = False
     while Kondisia == False:
       print('Input nilai a!')
       str_a = input('a = ')
       if "/" in str_a and str_a.count("/")==1:
         bila = str_a.split("/")

         try:
           # mengubah kedua bilangan menjadi float
           bila1 = float(bila[0])
           bila2 = float(bila[1])
           # mencetak hasil pembagian kedua bilangan
           a = bila1 / bila2
           Kondisia = True

         except ValueError:
           print('Input a harus bilangan!')

       else:
         if str_a.count("/")==0:
           try:
             a = float(str_a)
             Kondisia = True
           except ValueError:
             print('Input a harus bilangan!')
         else:
           print('Input a harus bilangan!')

     Kondisib = False
     while Kondisib == False:
       print('')
       print('Input nilai b!')
       str_b = input('b = ')
       if "/" in str_b and str_b.count("/")==1:
         bilb = str_b.split("/")

         try:
           # mengubah kedua bilangan menjadi float
           bilb1 = float(bilb[0])
           bilb2 = float(bilb[1])
           # mencetak hasil pembagian kedua bilangan
           b = bilb1 / bilb2
           Kondisib = True

         except ValueError:
           print('Input b harus bilangan!')

       else:
         if str_b.count("/")==0:
           try:
             b = float(str_b)
             Kondisib = True
           except ValueError:
             print('Input b harus bilangan')
         else:
           print('Input b harus bilangan')

     if a == 0 and b == 0:
       print('')
       print('Nilai a dan b tidak boleh 0 secara bersamaan!')
       print('Input kembali nilai a dan b!')
       print('')
     else:
       Kondisiab = True

    Kondisic = False
    while Kondisic == False:
       print('')
       print('Input nilai c!')
       str_c = input('c = ')
       if "/" in str_c and str_c.count("/")==1:
         bilc = str_c.split("/")

         try:
           # mengubah kedua bilangan menjadi float
           bilc1 = float(bilc[0])
           bilc2 = float(bilc[1])
           # mencetak hasil pembagian kedua bilangan
           c = bilc1 / bilc2
           Kondisic = True

         except ValueError:
           print('Input c harus bilangan!')

       else:
         if str_c.count("/")==0:
           try:
             c = float(str_c)
             Kondisic = True
           except ValueError:
             print('Input c harus bilangan!')
         else:
           print('Input c harus bilangan!')

    # Print Persamaan 1
    print('')

    # Ketika a<0
    if a < 0:
      if a!=-1:
        if b<0:
          if b!=-1: #a<0, a!=-1, b<0, b!=-1
            print('Persamaan 1: ', a, 'x ', b, 'y = ', c, sep='')
          else: #a<0, a!=-1, b<0, b=-1
            print('Persamaan 1: ', a, 'x - y = ', c, sep='')
        else:
          if b==0: #a<0, a!=-1, b=0
            print('Persamaan 1: ', a, 'x = ', c, sep='')
          else:
            if b==1: #a<0, a!=-1, b=1
               print('Persamaan 1: ', a, 'x + y = ', c, sep='')
            else: #a<0, a!=-1, b>0, b!=1
               print('Persamaan 1: ', a, 'x + ', b, 'y = ', c, sep='')

      if a==-1:
        if b<0:
          if b!=-1: #a=-1, b<0, b!=-1
            print('Persamaan 1: -x ', b, 'y = ', c, sep='')
          else: #a=-1, b<0, b=-1
            print('Persamaan 1: -x - y = ', c, sep='')
        else:
          if b==0: #a=-1, b=0
            print('Persamaan 1: -x = ', c, sep='')
          else:
            if b==1: #a=-1, b=1
               print('Persamaan 1: -x + y = ', c, sep='')
            else: #a=-1, b>0, b!=1
               print('Persamaan 1: -x + ', b, 'y = ', c, sep='')

    # Ketika a=0
    if a==0:
      if b<0:
        if b!=-1: #a=0, b<0, b!=-1
            print('Persamaan 1: ', b, 'y = ', c, sep='')
        else: #a=0, b<0, b=-1
            print('Persamaan 1: - y = ', c, sep='')
      else:
        if b==0: #a=0, b=0
            print('Persamaan 1: ', c, sep='')
        else:
          if b==1: #a=0, b=1
               print('Persamaan 1: y = ', c, sep='')
          else: #a=0, b>0, b!=1
               print('Persamaan 1: ', b, 'y = ', c, sep='')


    # Ketika a>0
    if a > 0:
      if a!=1:
        if b<0:
         if b!=-1: #a>0, b<0, b!=-1
            print('Persamaan 1: ', a, 'x', b, 'y = ', c, sep='')
         else: #a>0, b<0, b=-1
            print('Persamaan 1: ', a, 'x', '-y = ', c, sep='')
        else:
         if b==0: #a>0, b=0
            print('Persamaan 1: ', a, 'x = ', c, sep='')
         else:
          if b==1: #a>0, b=1
               print('Persamaan 1: ', a, 'x + y = ', c, sep='')
          else: #a>0, b>0, b!=1
               print('Persamaan 1: ', a, 'x + ', b, 'y = ', c, sep='')

    if a==1:
      if b<0:
         if b!=-1: #a=1, b<0, b!=-1
            print('Persamaan 1: x', b, 'y = ', c, sep='')
         else: #a=1, b<0, b=-1
            print('Persamaan 1: x - y = ', c, sep='')
      else:
         if b==0: #a=1, b=0
            print('Persamaan 1: x = ', c, sep='')
         else:
          if b==1: #a=1, b=1
               print('Persamaan 1: x + y = ', c, sep='')
          else: #a=1, b>0, b!=1
               print('Persamaan 1: x + ', b, 'y = ', c, sep='')

    time.sleep(10)
    print('\n')

    #Masukan Persamaan 2
    print('Ingat kembali dengan Bentuk Umum SPLDV')
    time.sleep(5)
    print('')
    print('Persamaan 1: ax + by = c')
    print('Persamaan 2: px + qy = r')
    time.sleep(10)
    print('')

    print('Dari persamaan 2, inputkanlah nilai-nilai koefisien (p dan q) dan konstantanya r)!')
    time.sleep(3)
    print('Ingat bahwa: Nilai p',"\u2260",'0 atau q',"\u2260",'0')
    time.sleep(3)
    print('Jika nilai p, q, atau r adalah bilangan pecahan')
    time.sleep(3)
    print('Teman-teman bisa inputkan dalam bentuk desimalnya')
    time.sleep(3)
    print('atau diinputkan dalam bentuk pecahan "m/n"')
    time.sleep(10)
    print('')

    Kondisipq = False
    while Kondisipq == False:

     Kondisip = False
     while Kondisip == False:
       print('Input nilai p!')
       str_p = input('p = ')
       if "/" in str_p and str_p.count("/")==1:
         bilp = str_p.split("/")

         try:
           # mengubah kedua bilangan menjadi float
           bilp1 = float(bilp[0])
           bilp2 = float(bilp[1])
           # mencetak hasil pembagian kedua bilangan
           p = bilp1 / bilp2
           Kondisip = True

         except ValueError:
           print('Input p harus bilangan!')

       else:
         if str_p.count("/")==0:
           try:
             p = float(str_p)
             Kondisip = True
           except ValueError:
             print('Input p harus bilangan!')
         else:
           print('Input p harus bilangan!')

     Kondisiq = False
     while Kondisiq == False:
       print('')
       print('Input nilai q!')
       str_q = input('q = ')
       if "/" in str_q and str_q.count("/")==1:
         bilq = str_q.split("/")

         try:
           # mengubah kedua bilangan menjadi float
           bilq1 = float(bilq[0])
           bilq2 = float(bilq[1])
           # mencetak hasil pembagian kedua bilangan
           q = bilq1 / bilq2
           Kondisiq = True

         except ValueError:
           print('Input q harus bilangan!')

       else:
         if str_q.count("/")==0:
           try:
             q = float(str_q)
             Kondisiq = True
           except ValueError:
             print('Input q harus bilangan!')
         else:
           print('Input q harus bilangan!')

     if p == 0 and q == 0:
       print('')
       print('Nilai p dan q tidak boleh 0 secara bersamaan!')
       print('Input kembali nilai p dan q!')
       print('')
     else:
       Kondisipq = True

    Kondisir = False
    while Kondisir == False:
       print('')
       print('Input nilai r!')
       str_r = input('r = ')
       if "/" in str_r and str_r.count("/")==1:
         bilr = str_r.split("/")

         try:
           # mengubah kedua bilangan menjadi float
           bilr1 = float(bilr[0])
           bilr2 = float(bilr[1])
           # mencetak hasil pembagian kedua bilangan
           r = bilr1 / bilr2
           Kondisir = True

         except ValueError:
           print('Input r harus bilangan!')

       else:
         if str_r.count("/")==0:
           try:
             r = float(str_r)
             Kondisir = True
           except ValueError:
             print('Input r harus bilangan!')
         else:
           print('Input r harus bilangan!')

    # Print Persamaan 2
    print('')

    # Ketika p<0
    if p < 0:
      if p!=-1:
        if q<0:
          if q!=-1: #p<0, p!=-1, q<0, q!=-1
            print('Persamaan 2: ', p, 'x ', q, 'y = ', r, sep='')
          else: #p<0, p!=-1, q<0, q=-1
            print('Persamaan 2: ', p, 'x - y = ', r, sep='')
        else:
          if q==0: #p<0, p!=-1, q=0
            print('Persamaan 2: ', p, 'x = ', r, sep='')
          else:
            if b==1: #p<0, p!=-1, q=1
               print('Persamaan 2: ', p, 'x + y = ', r, sep='')
            else: #p<0, p!=-1, q>0, q!=1
               print('Persamaan 2: ', p, 'x + ', q, 'y = ', r, sep='')

      if p==-1:
        if q<0:
          if q!=-1: #p=-1, q<0, q!=-1
            print('Persamaan 2: -x ', q, 'y = ', r, sep='')
          else: #p=-1, q<0, q=-1
            print('Persamaan 2: -x - y = ', r, sep='')
        else:
          if q==0: #a=-1, q=0
            print('Persamaan 2: -x = ', r, sep='')
          else:
            if q==1: #p=-1, q=1
               print('Persamaan 2: -x + y = ', r, sep='')
            else: #p=-1, q>0, q!=1
               print('Persamaan 2: -x + ', q, 'y = ', r, sep='')

    # Ketika a=0
    if p==0:
      if q<0:
        if q!=-1: #p=0, q<0, q!=-1
            print('Persamaan 2: ', q, 'y = ', r, sep='')
        else: #p=0, q<0, q=-1
            print('Persamaan 2: -y = ', r, sep='')
      else:
        if q==0: #p=0, q=0
            print('Persamaan 2: ', r, sep='')
        else:
          if q==1: #p=0, q=1
               print('Persamaan 2: y = ', r, sep='')
          else: #p=0, q>0, q!=1
               print('Persamaan 2: ', q, 'y = ', r, sep='')


    # Ketika p>0
    if p > 0:
      if p!=1:
        if q<0:
         if q!=-1: #p>0, q<0, q!=-1
            print('Persamaan 2: ', p, 'x', q, 'y = ', r, sep='')
         else: #p>0, q<0, q=-1
            print('Persamaan 2: ', p, 'x', '-y = ', r, sep='')
        else:
         if q==0: #p>0, q=0
            print('Persamaan 2: ', p, 'x = ', r, sep='')
         else:
          if q==1: #q>0, q=1
               print('Persamaan 2: ', p, 'x + y = ', r, sep='')
          else: #a>0, q>0, q!=1
               print('Persamaan 2: ', p, 'x + ', q, 'y = ', r, sep='')

    if p==1:
      if q<0:
         if q!=-1: #p=1, q<0, q!=-1
            print('Persamaan 2: x', q, 'y = ', r, sep='')
         else: #p=1, q<0, q=-1
            print('Persamaan 2: x - y = ', r, sep='')
      else:
         if b==0: #p=1, q=0
            print('Persamaan 2: x = ', r, sep='')
         else:
          if q==1: #p=1, q=1
               print('Persamaan 2: x + y = ', r, sep='')
          else: #p=1, q>0, q!=1
               print('Persamaan 2: x + ', q, 'y = ', r, sep='')

    time.sleep(10)


    #Cek solusi SPL
    if a*q - p*b == 0:
      if (a/p) == (c/r):
         print('')
         print("SPLDV tersebut memiliki banyak solusi")
      else:
         print('')
         print("SPLDV tersebut tidak memiliki solusi")

    else:
      # Hitung nilai y dan x
      x = (c*q-b*r)/(a*q-b*p)
      x = round(x, 2)
      y = (a*r-c*p)/(a*q-b*p)
      y = round(y, 2)
      print('')
      print('Solusi SPLDV tersebut adalah')
      print('x=',x,'dan y=',y)
      time.sleep(10)

    print('')
    print('Bagaimana apakah jawaban yang diberikan')
    print('sudah sama dengan kunci jawaban punya teman-teman?')
    print('')
    time.sleep(5)
    print('Jika tidak sama, cek kembali kunci jawaban')
    print('atau coba teman-teman inputkan kembali SPLDV nya!')
    time.sleep(10)
    print('')
    print('Teman-teman akan diarahkan dulu ke Menu Utama ya!')
    print('')
    time.sleep(5)
    MenuUtama()
````
